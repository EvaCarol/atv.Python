```python
#1 questão: Faça um programa que identifique se a letra digitada é impar ou par?

letra = input("Digite uma letra: ").lower()

# Verifica se a letra digitada é uma vogal
if letra in 'aeiou':
    print("Sua letra é uma vogal.")
else:
    print("Sua letra é uma consoante.")

#2 questão: Faça um programa que pergunte o preço de três produtos e informe qual produto você deve comprar, sabendo que a decisão é sempre pelo mais barato.

preco1 = float(input("Digite o valor do primeiro produto: "))
preco2 = float(input("Digite o valor do segundo produto: "))
preco3 = float(input("Digite o valor do terceiro produto: "))

# Determina o menor preço e recomenda a compra do produto correspondente
menor_preco = min(preco1, preco2, preco3)
if menor_preco == preco1:
    print(f'Compre o produto 1 pelo valor de R${preco1:.2f}')
elif menor_preco == preco2:
    print(f'Compre o produto 2 pelo valor de R${preco2:.2f}')
else:
    print(f'Compre o produto 3 pelo valor de R${preco3:.2f}')

#3 questão: Faça um programa que leia três números e mostre-os em ordem crescente     

num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
num3 = float(input("Digite mais um número: "))

lista_numeros = [num1, num2, num3]
lista_numeros.sort()
menor_num, medio_num, maior_num = lista_numeros

print(f'Os números em ordem crescente são: {menor_num}, {medio_num}, {maior_num}')  

#4 questão: Faça um programa que pergunte em que turno você estuda. Peça para digitar M-matutino e V-verpertino ou N-noturno. Imprima a mensagem "Bom dia!","Boa tarde!" ou "Boa noite" ou "valor inválido,conforme o caso"

periodo_estudo = input("Informe o período de estudos: M para manhã, T para tarde, N para noite: ").upper()

match periodo_estudo:
    case "M":
        print("Bom dia!")
    case "T":
        print("Boa Tarde!")
    case "N":
        print("Boa noite!")
    case _:
        print("Opção inválida")

#5 Faça um programa que leia um número e exiba o dia correspondente da semana(1-Domingo, 2-Segunda, etc). Se for digitado outro valor deve retornar uma mensagem informando o valor inválido.

dias_da_semana = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']

def obter_dia(numero):
    if 1 <= numero <= 7:
        return dias_da_semana[numero - 1]
    else:
        return 'Valor inválido. Por favor, insira um número entre 1 e 7.'

numero = int(input('Digite um número (1 a 7) para obter o dia correspondente da semana: '))

print(obter_dia(numero))   

#6 Faça um programa que lê duas notas parciais obtidas por um aluno numa disciplina ao longo de um semestre, e calcule sua média. A atribuição de conceitos obedece às regras:
#O programa deve mostrar na tela as notas, a média, o conceito correspondente e a mensagem 'APROVADO' se o conceito for A,B ou C ou 'REPROVADO' se o conceito for D ou E.

nota1, nota2 = float(input('Digite a primeira nota: ')), float(input('Digite a segunda nota: '))
media = (nota1 + nota2) / 2
conceito = 'A' if media >= 9 else 'B' if media >= 7.5 else 'C' if media >= 6 else 'D' if media >= 4 else 'E'
status = 'APROVADO' if conceito in 'ABC' else 'REPROVADO'

print(f'Notas: {nota1}, {nota2}\nMédia: {media}\nConceito: {conceito}\n{status}')

#7 questão: Faça um programa que peça um número correspondente a um determinado ano e em seguida informe se este ano é ou não bissexto.
#Para o ano ser bissexto ele devr ser divisivel por 4,100,400

def eh_bissexto(ano):
    return ano % 4 == 0 and (ano % 100 != 0 or ano % 400 == 0)
ano = int(input('Digite um ano para verificar se é bissexto: '))

if eh_bissexto(ano):
    print(f'O ano {ano} é bissexto.')
else:
    print(f'O ano {ano} não é bissexto.')

#8 Faça um programa que leia 2 números e em seguida pergunte ao usuário qual operação básica(soma,subtração,divisão,multiplicação e exponenciação)ele deseja realizar. O resultado da operação deve ser acompanhado de uma frase que diga se o número é : par ou impar; positivo ou negativo; inteiro ou decimal;    

def descrever_numero(numero):
    descricao = 'par' if numero % 2 == 0 else 'ímpar'
    descricao += '; positivo' if numero > 0 else '; negativo' if numero < 0 else '; zero'
    descricao += '; inteiro' if numero.is_integer() else '; decimal'
    return descricao

num1 = float(input('Digite o primeiro número: '))
num2 = float(input('Digite o segundo número: '))

operacao = input('Qual operação você deseja realizar? (soma, subtração, divisão, multiplicação, exponenciação): ').lower()

if operacao == 'soma':
    resultado = num1 + num2
elif operacao == 'subtração':
    resultado = num1 - num2
elif operacao == 'divisão':
    resultado = num1 / num2
elif operacao == 'multiplicação':
    resultado = num1 * num2
elif operacao == 'exponenciação':
    resultado = num1 ** num2
else:
    print('Operação inválida.')
    exit()

print(f'Resultado: {resultado}')
print(descrever_numero(resultado))

#9 Faça um porgrama que peça a idade e exiba a informação. Os valores aceitáveis devem estar entre 0 e 150. Valores fora desse limite não poderão ser aceitos.

idade = int(input('Digite sua idade: '))

if 0 <= idade <= 150:
    print(f'A idade informada é: {idade}')
else:
    print('Idade inválida. Por favor, insira um valor entre 0 e 150.')

#10 Utilizando estruturas de repetição e listas(com ou sem .append), faça um programa que leia 5 números e informe a soma e a média dos números.

numeros = []
soma = 0

for i in range(5):
    numero = float(input(f'Digite o número {i+1}: '))
    numeros.append(numero)
    soma += numero

media = soma / len(numeros)

print(f'Números informados: {numeros}')
print(f'Soma dos números: {soma}')
print(f'Média dos números: {media:.2f}')

#11 Faça um programa que peça um número inteiro e determine se ele é ou não primo. Um número primo é aquele que é divisível somente ele mesmo e por 1.

def eh_primo(numero):
    if numero <= 1:
        return False
    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            return False
    return True

num = int(input('Digite um número inteiro: '))

if eh_primo(num):
    print(f'O número {num} é primo.')
else:
    print(f'O número {num} não é primo.')

#12 Faça um programa que leia 5 temperaturas, e informe ao final a menor e a maior 

temperaturas = []

for i in range(5):
    temp = float(input(f'Digite a temperatura {i+1}: '))
    temperaturas.append(temp)

menor = min(temperaturas)
maior = max(temperaturas)

print(f'A menor temperatura registrada foi: {menor}°C')
print(f'A maior temperatura registrada foi: {maior}°C')

#13 INSS

saldo_medio = 400.80

if saldo_medio <= 200:
    taxa = 0
elif saldo_medio <= 400:
    taxa = 0.2
elif saldo_medio <= 600:
    taxa = 0.3
else:
    taxa = 0.4

credito = saldo_medio * taxa

if taxa == 0:
    print("Você não possui crédito.")
else:
    print(f"Seu saldo médio é: R$ {saldo_medio:.2f}, e o valor do seu crédito é: R$ {credito:.2f}")

#14 INSS

nome_usuario = input("Por favor, informe seu nome: ")
idade_usuario = int(input("Qual é a sua idade atual? "))
tempo_restante_aposentadoria = 65 - idade_usuario

info_aposentadoria = f"{nome_usuario}, faltam {tempo_restante_aposentadoria} anos para você se aposentar."
print(info_aposentadoria)    

#15 Calcular folha de pagamento

valor_hora = float(input("Digite o valor da sua hora de trabalho: "))
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas no mês: "))

salario_bruto = valor_hora * horas_trabalhadas

aliquotas_ir = [
    (4664.68, 0.275),
    (3751.05, 0.225),
    (2826.65, 0.15),
    (2112.00, 0.075),
    (0, 0)
]
desconto_ir = next(aliquota for limite, aliquota in aliquotas_ir if salario_bruto > limite) * salario_bruto
desconto_sindicato = salario_bruto * 0.03
fgts = salario_bruto * 0.11

salario_liquido = salario_bruto - desconto_ir - desconto_sindicato

print(f"Salário Bruto: R$ {salario_bruto:.2f}")
print(f"Desconto do Imposto de Renda: R$ {desconto_ir:.2f}")
print(f"Desconto do Sindicato: R$ {desconto_sindicato:.2f}")
print(f"FGTS: R$ {fgts:.2f}")
print(f"Salário Líquido: R$ {salario_liquido:.2f}")
```
